# Salary Prediction Model using Linear Regression
## A Machine Learning Project for Salary Estimation

This project implements a linear regression model to predict salaries based on years of experience. The model achieves an accuracy of 95.96% and can be valuable for:
- HR professionals in salary negotiations
- Job seekers in salary expectations
- Businesses in compensation planning

## Project Overview
- **Objective**: Develop a predictive model for salary estimation based on experience
- **Model**: Linear Regression with Gradient Descent implementation
- **Accuracy**: 95.96%
- **Features**: Years of Experience
- **Target**: Salary

## Implementation Steps
1. Data Loading and Preprocessing
2. Exploratory Data Analysis

4. Model Evaluation
5. Salary Prediction

### 1. Import Required Libraries

```python
# Data manipulation and analysis
import pandas as pd
import numpy as np

# Visualization libraries
import matplotlib.pyplot as plt
import seaborn as sns

# Machine Learning libraries
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

import warnings
warnings.filterwarnings("ignore")
```

### 2. Load and Explore Training Data

```python
# Load the training dataset
df = pd.read_csv(r"data\Salary_Train_data.csv")
print("Training Data Shape:", df.shape)
print("\nFirst 5 rows of the dataset:")
df.head()
```

### 3. Data Visualization

```python
# Create a scatter plot with color gradient
x = df['YearsExperience']
y = df['Salary']

plt.figure(figsize=(10, 6))
scatter = plt.scatter(x, y, c=range(len(x)), cmap='viridis')
plt.colorbar(scatter, label='Data Point Index')
plt.title("Experience vs Salary Distribution", fontsize=15, pad=20)
plt.grid(True, linestyle='--', alpha=0.5)
plt.xlabel("Years of Experience", fontsize=12)
plt.ylabel("Salary", fontsize=12)
plt.show()

# Create a bar plot for salary distribution
plt.figure(figsize=(12, 6))
sns.barplot(x="YearsExperience", y="Salary", data=df)
plt.title("Salary Distribution by Years of Experience", fontsize=15, pad=20)
plt.grid(alpha=0.5)
plt.xlabel("Years of Experience", fontsize=12)
plt.ylabel("Salary", fontsize=12)
plt.xticks(rotation=45)
plt.show()
```

### 4. Prepare Data for Modeling

```python
# Separate features (X) and target variable (y)
X = df[["YearsExperience"]]
y = df["Salary"]
```

### 5. Split Data into Training and Test Sets

```python
# Split the data with 80% training and 20% testing
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.80, random_state=5)
print(f"Training set size: {len(X_train)}")
print(f"Testing set size: {len(X_test)}")
```

### 6. Initialize and Train the Model

```python
# Initialize Linear Regression model
model = LinearRegression()
```

```python
# Train the model
model.fit(X_train.values, y_train.values)
print("Model training completed successfully!")
```

### 7. Model Parameters

```python
# Extract model parameters
m = model.coef_
c = model.intercept_
print(f"Slope (m) = {m[0]:.2f}")
print(f"Intercept (c) = {c:.2f}")
print("\nEquation: Salary = {:.2f}  YearsExperience + {:.2f}".format(m[0], c))
```

### 8. Make Predictions

```python
# Generate predictions on test set
y_pred = model.predict(X_test)
print("Sample predictions:")
for i in range(5):
    print(f"Experience: {X_test.iloc[i]['YearsExperience']:.1f} years, Predicted Salary: ${y_pred[i]:.2f}")
```

### 9. Model Evaluation
The models accuracy is calculated using R score:
- R = SSR / SST
where:
- SSR (Sum of Squares Regression) = (y_pred - y_mean)
- SSE (Sum of Squares Error) = (y_true - y_pred)
- SST (Total Sum of Squares) = SSR + SSE

```python
# Calculate model accuracy
accuracy = model.score(X_test, y_test)
print(f"Model Accuracy: {accuracy*100:.2f}%")
```

### 10. Visualize Model Results

```python
df["y_pred"] = model.predict(df[["YearsExperience"]])
# Plot actual vs predicted values
plt.figure(figsize=(12, 6))
sns.scatterplot(x="YearsExperience", y="Salary", data=df, color="red", alpha=0.5, label="Actual")
sns.lineplot(x=df["YearsExperience"], y=df["y_pred"], color="green", label="Predicted")
plt.title("Salary Prediction Model: Actual vs Predicted", fontsize=15, pad=20)
plt.xlabel("Years of Experience", fontsize=12)
plt.ylabel("Salary", fontsize=12)
plt.grid(True, alpha=0.3)
plt.legend()
plt.show()
```

### 11. Predict Salaries for New Data

```python
# Load test data
test_data = pd.read_csv(r"data\Salary_Test_data.csv")
print("Test Data Shape:", test_data.shape)
print("\nFirst 5 rows of test data:")
test_data.head()
```

```python
# Generate predictions for test data
test_data["Predicted_Salary"] = model.predict(test_data)
print("\nPredictions for test data:")
test_data.head()
```

## Custom Linear Regression Implementation
This section demonstrates the implementation of Linear Regression from scratch using Gradient Descent.

### 1. Import Required Libraries

```python
# Data manipulation and analysis
import pandas as pd
import numpy as np

# Visualization libraries
import matplotlib.pyplot as plt
import seaborn as sns
```

### 2. Load Training Data

```python
# Load the training dataset
df = pd.read_csv(r"data\Salary_Train_data.csv")
print("Training Data Shape:", df.shape)
print("\nFirst 5 rows of the dataset:")
df.head()
```

### 3. Data Visualization

```python
# Create a scatter plot with color gradient
x = df['YearsExperience']
y = df['Salary']

plt.figure(figsize=(12, 6))
scatter = plt.scatter(x, y, c=range(len(x)), cmap='hsv')
plt.colorbar(scatter, label='Data Point Index')
plt.title("Experience vs Salary Distribution", fontsize=15, pad=20)
plt.grid(True, linestyle='--', alpha=0.5)
plt.xlabel("Years of Experience", fontsize=12)
plt.ylabel("Salary", fontsize=12)
plt.show()
```

### 4. Prepare Data for Modeling

```python
# Separate features and target variable
y = df["Salary"]
x = df["YearsExperience"]
```

### 5. Implement Gradient Descent

```python
def gradient_descent(x, y):
    """
    Implement Linear Regression using Gradient Descent
    
    Parameters:
    x: Feature values (Years of Experience)
    y: Target values (Salary)
    
    Returns:
    m: Slope of the line
    c: Y-intercept
    """
    global m, c
    m = 0 
    c = 0
    n = len(x)
    learning_rate = 0.026
    previous_mse = 0
    
    plt.figure(figsize=(12, 6))
    
    while True:
        # Calculate predictions
        y_pred = m * x + c
        
        # Plot current line
        plt.plot(x, y_pred, color="green", alpha=0.3)
        
        # Calculate Mean Squared Error
        mse = (1/n) * sum([i**2 for i in (y-y_pred)])
        
        # Calculate gradients
        gradient_m = -(2/n) * sum(x * (y-y_pred))
        gradient_c = -(2/n) * sum(y-y_pred)
        
        # Update parameters
        m = m - learning_rate * gradient_m
        c = c - learning_rate * gradient_c
        
        # Check convergence
        if mse == previous_mse:
            plt.scatter(x, y, c="blue", alpha=0.5, label="Data Points")
            plt.plot(x, m*x+c, color="red", label="Final Line")
            plt.title("Gradient Descent: Model Fitting", fontsize=15, pad=20)
            plt.xlabel("Years of Experience", fontsize=12)
            plt.ylabel("Salary", fontsize=12)
            plt.grid(True, alpha=0.3)
            plt.legend()
            print("Model converged successfully!")
            break
        else:
            previous_mse = mse
    
    plt.show()
    return m, c
```

### 6. Train the Model

```python
# Train the model using gradient descent
m, c = gradient_descent(x, y)
```

### 7. Model Parameters

```python
print(f"Slope (m) = {m:.2f}")
print(f"Intercept (c) = {c:.2f}")
print("\nEquation: Salary = {:.2f}  YearsExperience + {:.2f}".format(m, c))
```

### 8. Generate Predictions

```python
# Calculate predictions
y_predict = m * x + c
print("\nSample predictions:")
for i in range(5):
    print(f"Experience: {x.iloc[i]:.1f} years, Predicted Salary: ${y_predict.iloc[i]:.2f}")
```

### 9. Model Evaluation

```python
# Calculate R score components
y = np.array(y)
y_predict = np.array([m*x+c])

# Calculate Sum of Squares Regression (SSR)
sum_squares_regression = sum(y_predict-y.mean())**2
SSR = sum_squares_regression.sum()

# Calculate Sum of Squares Error (SSE)
sum_squares_error = sum((y-y_predict)**2)
SSE = sum_squares_error.sum()

# Calculate Total Sum of Squares (SST)
SST = SSR + SSE

# Calculate R score
R_squared = SSR / SST
print(f"Model Accuracy (R Score): {R_squared*100:.2f}%")
```

### 10. Visualize Results

```python
# Plot actual vs predicted values
plt.figure(figsize=(12, 6))
sns.scatterplot(x="YearsExperience", y="Salary", data=df, color="red", alpha=0.5, label="Actual")
sns.lineplot(x=df["YearsExperience"], y=m*df["YearsExperience"]+c, color="green", label="Predicted")
plt.title("Custom Linear Regression: Actual vs Predicted", fontsize=15, pad=20)
plt.xlabel("Years of Experience", fontsize=12)
plt.ylabel("Salary", fontsize=12)
plt.grid(True, alpha=0.3)
plt.legend()
plt.show()
```

### 11. Predict Salaries for New Data

```python
# Load test data
test_data = pd.read_csv(r"data\Salary_Test_data.csv")
print("Test Data Shape:", test_data.shape)
print("\nFirst 5 rows of test data:")
test_data.head()
```

```python
# Generate predictions for test data
test_data["Predicted_Salary"] = test_data["YearsExperience"].apply(lambda x: m * x + c)
print("\nPredictions for test data:")
test_data.head()
```
